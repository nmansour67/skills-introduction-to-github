```python
# ===========================
# STEP 1.5: DATASET UPLOAD & VALIDATION
# ===========================

import os # Import the os module

print("\n" + "=" * 60)
print("DATASET UPLOAD")
print("=" * 60)
print("""
This Digital Twin simulation can be enhanced with real patient data from your
organization. The dataset should contain historical RAI-HC assessments and
clinical outcomes.

ðŸ“Š OPTION 1: Upload your own dataset (CSV format) - Click button below
ðŸ“‹ OPTION 2: Use the provided template dataset - Leave empty and run next cell

Required columns for uploaded dataset:
  - patient_id: Unique identifier
  - age: Patient age in years
  - heart_failure: 1 if diagnosed, 0 otherwise
  - diabetes: 1 if diagnosed, 0 otherwise
  - ckd: 1 if chronic kidney disease, 0 otherwise
  - hypertension: 1 if diagnosed, 0 otherwise
  - bp_systolic: Systolic blood pressure (mmHg)
  - glucose: Blood glucose level (mg/dL)
  - edema_scale: Pitting edema scale (0-4)
  - falls_6months: Number of falls in past 6 months
  - medication_count: Number of prescribed medications
  - hospitalized: 1 if hospitalized within 30 days, 0 otherwise
  - hospitalization_day: Day of hospitalization (or NaN if not hospitalized)
  - length_of_stay: Hospital LOS in days (or 0 if not hospitalized)
  - aki_developed: 1 if acute kidney injury occurred, 0 otherwise
  - readmitted_30d: 1 if readmitted within 30 days, 0 otherwise

""")

# Import file upload widget for Google Colab
try:
    from google.colab import files
    import io
    IN_COLAB = True
except ImportError:
    IN_COLAB = False
    print("âš  Not running in Google Colab. File upload widget unavailable.")

def generate_template_dataset(n_patients=200):
    """
    Generate synthetic RAI-HC dataset for demonstration purposes

    Args:
        n_patients: Number of synthetic patient records to generate

    Returns:
        pandas DataFrame with realistic clinical data
    """
    np.random.seed(42)  # For reproducibility

    # Generate patient demographics
    patient_ids = [f"PT{str(i).zfill(5)}" for i in range(1, n_patients + 1)]
    ages = np.random.normal(75, 8, n_patients).clip(50, 95).astype(int)

    # Generate comorbidities (correlated with age)
    heart_failure = (np.random.random(n_patients) < (0.15 + (ages - 50) / 200)).astype(int)
    diabetes = (np.random.random(n_patients) < 0.35).astype(int)
    ckd = (np.random.random(n_patients) < (0.20 + heart_failure * 0.15)).astype(int)
    hypertension = (np.random.random(n_patients) < 0.60).astype(int)

    # Generate clinical measurements
    bp_systolic = np.random.normal(145, 20, n_patients).clip(100, 200).astype(int)
    bp_systolic += hypertension * 15  # Higher BP if hypertensive

    glucose = np.random.normal(150, 40, n_patients).clip(70, 400).astype(int)
    glucose += diabetes * 50  # Higher glucose if diabetic

    edema_scale = np.random.choice([0, 1, 2, 3, 4], n_patients,
                                    p=[0.50, 0.25, 0.15, 0.07, 0.03])
    edema_scale += (heart_failure * np.random.choice([0, 1, 2], n_patients,
                                                      p=[0.3, 0.5, 0.2]))
    edema_scale = edema_scale.clip(0, 4)

    falls_6months = np.random.poisson(0.5, n_patients)

    medication_count = np.random.poisson(5, n_patients).clip(0, 20)
    medication_count += heart_failure * 2 + diabetes * 1 + ckd * 1

    # Calculate risk scores (similar to simulation logic)
    risk_scores = (
        (ages > 75) * 2 +
        heart_failure * 3 +
        diabetes * 2 +
        ckd * 2 +
        hypertension * 1 +
        (bp_systolic > 160) * 1 +
        (glucose > 200) * 1 +
        (edema_scale >= 2) * 2 +
        (falls_6months > 0) * 1
    )

    # Generate outcomes based on risk scores
    hosp_probability = (risk_scores / 20).clip(0.05, 0.65)
    hospitalized = (np.random.random(n_patients) < hosp_probability).astype(int)

    hospitalization_day = np.where(
        hospitalized == 1,
        np.random.choice([7, 8, 9, 10], n_patients, p=[0.25, 0.40, 0.20, 0.15]),
        np.nan
    )

    length_of_stay = np.where(
        hospitalized == 1,
        np.random.choice([2, 3, 4, 5, 6, 7], n_patients,
                        p=[0.15, 0.25, 0.25, 0.20, 0.10, 0.05]),
        0
    )

    # AKI more likely with delayed hospitalization and CKD
    aki_probability = (
        hospitalized * 0.15 +
        (hospitalization_day == 8) * 0.20 +
        ckd * 0.15
    )
    aki_developed = (np.random.random(n_patients) < aki_probability).astype(int)

    # Readmission correlated with multiple factors
    readmit_probability = (
        hospitalized * 0.15 +
        aki_developed * 0.10 +
        (length_of_stay > 5) * 0.08 +
        (medication_count > 8) * 0.05
    )
    readmitted_30d = (np.random.random(n_patients) < readmit_probability).astype(int)

    # Assemble dataset
    template_data = pd.DataFrame({
        'patient_id': patient_ids,
        'age': ages,
        'heart_failure': heart_failure,
        'diabetes': diabetes,
        'ckd': ckd,
        'hypertension': hypertension,
        'bp_systolic': bp_systolic,
        'glucose': glucose,
        'edema_scale': edema_scale,
        'falls_6months': falls_6months,
        'medication_count': medication_count,
        'hospitalized': hospitalized,
        'hospitalization_day': hospitalization_day,
        'length_of_stay': length_of_stay,
        'aki_developed': aki_developed,
        'readmitted_30d': readmitted_30d
    })

    return template_data

def validate_dataset(df):
    """
    Validate uploaded dataset structure and content

    Args:
        df: pandas DataFrame to validate

    Returns:
        tuple: (is_valid: bool, error_messages: list)
    """
    errors = []

    # Check required columns
    required_cols = [
        'patient_id', 'age', 'heart_failure', 'diabetes', 'ckd',
        'hypertension', 'bp_systolic', 'glucose', 'edema_scale',
        'falls_6months', 'medication_count', 'hospitalized',
        'hospitalization_day', 'length_of_stay', 'aki_developed',
        'readmitted_30d'
    ]

    missing_cols = set(required_cols) - set(df.columns)
    if missing_cols:
        errors.append(f"Missing required columns: {', '.join(missing_cols)}")

    # Check data types and ranges
    if 'age' in df.columns:
        if df['age'].min() < 0 or df['age'].max() > 120:
            errors.append("Age values out of valid range (0-120)")

    if 'bp_systolic' in df.columns:
        if df['bp_systolic'].min() < 60 or df['bp_systolic'].max() > 250:
            errors.append("Blood pressure values out of valid range (60-250)")

    if 'edema_scale' in df.columns:
        if df['edema_scale'].min() < 0 or df['edema_scale'].max() > 4:
            errors.append("Edema scale must be 0-4")

    # Check binary columns
    binary_cols = ['heart_failure', 'diabetes', 'ckd', 'hypertension',
                   'hospitalized', 'aki_developed', 'readmitted_30d']
    for col in binary_cols:
        if col in df.columns:
            unique_vals = df[col].dropna().unique()
            if not set(unique_vals).issubset({0, 1}):
                errors.append(f"{col} must contain only 0 or 1 values")

    # Check minimum sample size
    if len(df) < 20:
        errors.append(f"Dataset too small ({len(df)} rows). Minimum 20 patients required.")

    is_valid = len(errors) == 0
    return is_valid, errors

# Main upload logic
patient_dataset = None

if IN_COLAB:
    print("=" * 60)
    print("ðŸ“ CLICK 'Choose Files' BUTTON BELOW TO UPLOAD YOUR CSV FILE")
    print("   (Or leave empty to use template dataset)")
    print("=" * 60)
    print()

    # This displays the actual file upload button
    uploaded = files.upload()

    if uploaded:
        # User uploaded a file
        filename = list(uploaded.keys())[0]
        print(f"\nâœ“ File '{filename}' received")

        try:
            # Read the uploaded CSV file
            patient_dataset = pd.read_csv(io.BytesIO(uploaded[filename]))
            print(f"âœ“ Dataset loaded: {len(patient_dataset)} patients, {len(patient_dataset.columns)} columns")

            # Validate the dataset
            is_valid, errors = validate_dataset(patient_dataset)

            if is_valid:
                print("âœ… Dataset validation PASSED - Ready for simulation!")
                print("\nðŸ“Š Dataset preview (first 5 rows):")
                print(patient_dataset.head())
            else:
                print("\nâŒ Dataset validation FAILED:")
                for error in errors:
                    print(f"   â€¢ {error}")
                print("\nâš ï¸  Falling back to template dataset...")
                patient_dataset = generate_template_dataset()

        except Exception as e:
            print(f"\nâŒ Error reading CSV file: {str(e)}")
            print("âš ï¸  Falling back to template dataset...")
            patient_dataset = generate_template_dataset()
    else:
        # No file uploaded - use template
        print("\nðŸ“‹ No file uploaded. Generating template dataset...")
        patient_dataset = generate_template_dataset(n_patients=200)
        print(f"âœ“ Template dataset created: {len(patient_dataset)} synthetic patients")
        print("\nðŸ“Š Dataset preview (first 5 rows):")
        print(patient_dataset.head())
else:
    # Not in Colab - automatically use template
    print("\nâš ï¸  Not running in Google Colab environment")
    print("ðŸ“‹ Generating template dataset...")
    patient_dataset = generate_template_dataset(n_patients=200)
    print(f"âœ“ Template dataset created: {len(patient_dataset)} patients")

# Display summary statistics
print("\n" + "=" * 60)
print("ðŸ“ˆ DATASET SUMMARY STATISTICS")
print("=" * 60)

summary_stats = {
    'Total Patients': len(patient_dataset),
    'Mean Age': f"{patient_dataset['age'].mean():.1f} years",
    'Age Range': f"{patient_dataset['age'].min()}-{patient_dataset['age'].max()} years",
    'Heart Failure Prevalence': f"{(patient_dataset['heart_failure'].sum() / len(patient_dataset) * 100):.1f}%",
    'Diabetes Prevalence': f"{(patient_dataset['diabetes'].sum() / len(patient_dataset) * 100):.1f}%",
    'CKD Prevalence': f"{(patient_dataset['ckd'].sum() / len(patient_dataset) * 100):.1f}%",
    'Hypertension Prevalence': f"{(patient_dataset['hypertension'].sum() / len(patient_dataset) * 100):.1f}%",
    'Mean BP Systolic': f"{patient_dataset['bp_systolic'].mean():.0f} mmHg",
    'Mean Glucose': f"{patient_dataset['glucose'].mean():.0f} mg/dL",
    'Hospitalization Rate': f"{(patient_dataset['hospitalized'].sum() / len(patient_dataset) * 100):.1f}%",
    'Mean LOS (if hospitalized)': f"{patient_dataset[patient_dataset['hospitalized'] == 1]['length_of_stay'].mean():.1f} days",
    'AKI Rate': f"{(patient_dataset['aki_developed'].sum() / len(patient_dataset) * 100):.1f}%",
    '30-day Readmission Rate': f"{(patient_dataset['readmitted_30d'].sum() / len(patient_dataset) * 100):.1f}%"
}

for key, value in summary_stats.items():
    print(f"{key:.<45} {value}")

print("=" * 60)

# Save template for user reference
template_path = '/mnt/user-data/outputs/template_dataset.csv'
# Create the directory if it doesn't exist
os.makedirs(os.path.dirname(template_path), exist_ok=True)
patient_dataset.to_csv(template_path, index=False)
print(f"\nðŸ’¾ Dataset saved to: {template_path}")
print("   (You can download this file to see the expected format)")

print("\nâœ… Dataset ready for Digital Twin simulation")
print("=" * 60)
```

```python
# ===========================
# STEP 2: DEFINE PATIENT PARAMETERS
# ===========================

class Patient:
    """Digital twin of Mr. Hanna"""

    def __init__(self, name, age, conditions):
        self.name = name
        self.age = age
        self.conditions = conditions
        self.history = []

    def get_baseline_risk(self):
        """Calculate baseline hospitalization risk using simplified scoring"""
        risk_score = 0

        # Age factor
        if self.age > 75:
            risk_score += 2
        elif self.age > 65:
            risk_score += 1

        # Comorbidity factors
        comorbidity_weights = {
            'heart_failure': 3,
            'diabetes': 2,
            'ckd': 2,
            'hypertension': 1
        }

        for condition, weight in comorbidity_weights.items():
            if condition in self.conditions:
                risk_score += weight

        return risk_score

    def add_event(self, day, event_type, details):
        """Log care pathway events"""
        self.history.append({
            'day': day,
            'event': event_type,
            'details': details
        })

# Create Mr. Hanna's digital twin
mr_hanna = Patient(
    name="Hassan Hanna",
    age=78,
    conditions=['heart_failure', 'diabetes', 'ckd', 'hypertension']
)

baseline_risk = mr_hanna.get_baseline_risk()
print(f"Patient: {mr_hanna.name}, Age: {mr_hanna.age}")
print(f"Baseline hospitalization risk score: {baseline_risk}/10")
print(f"Conditions: {', '.join(mr_hanna.conditions)}")
print("=" * 60)
```

```python
# ===========================
# STEP 3: DEFINE CARE PATHWAY STAGES
# ===========================

class CarePathway:
    """Simulates the care journey with decision points"""

    def __init__(self, patient):
        self.patient = patient
        self.current_day = 0
        self.hospitalized = False
        self.hospitalization_day = None

    def day1_assessment(self, escalation_protocol=False):
        """
        Day 1: Home care assessment

        Args:
            escalation_protocol: If True, uses risk algorithm
        """
        self.current_day = 1

        # Clinical findings
        bp_systolic = 165
        glucose = 245
        edema = 2  # pitting scale

        # Calculate clinical deterioration score
        deterioration_score = 0
        if bp_systolic > 160:
            deterioration_score += 1
        if glucose > 200:
            deterioration_score += 1
        if edema >= 2:
            deterioration_score += 2

        # CRITICAL DECISION POINT #1
        if escalation_protocol:
            # With algorithm: objective criteria trigger urgent alert
            combined_risk = self.patient.get_baseline_risk() + deterioration_score

            if combined_risk >= 8:
                urgency = "URGENT"
                gp_response_days = np.random.choice([1, 2], p=[0.7, 0.3])
            else:
                urgency = "ROUTINE"
                gp_response_days = np.random.choice([3, 7, 14], p=[0.3, 0.4, 0.3])
        else:
            # Without algorithm: subjective judgment
            urgency = "ROUTINE"
            gp_response_days = np.random.choice([7, 14], p=[0.4, 0.6])

        self.patient.add_event(
            day=1,
            event_type="Assessment",
            details=f"BP {bp_systolic}/95, Glucose {glucose}, Edema +{edema}. "
                   f"Tagged as {urgency}, GP follow-up in {gp_response_days} days"
        )

        return gp_response_days

    def day7_fall(self, weekend_protocol=False, family_education=False):
        """
        Day 7: Weekend fall incident

        Args:
            weekend_protocol: If True, uses standardized response
            family_education: If True, family knows warning signs
        """
        self.current_day = 7

        # CRITICAL DECISION POINT #2
        if weekend_protocol:
            # Protocol: "Any fall in CHF patient = ER evaluation"
            er_probability = 0.85
        elif family_education:
            # Family recognizes warning signs, escalates appropriately
            er_probability = 0.65
        else:
            # Advice: "Monitor overnight"
            er_probability = 0.15

        goes_to_er = np.random.random() < er_probability

        if goes_to_er:
            self.hospitalized = True
            self.hospitalization_day = 7
            self.patient.add_event(
                day=7,
                event_type="Fall â†’ ER",
                details="Family brought to ER immediately. Early intervention."
            )
        else:
            self.patient.add_event(
                day=7,
                event_type="Fall â†’ Monitor",
                details="Advised to monitor overnight. Condition worsens."
            )

        return goes_to_er

    def day8_crisis(self):
        """Day 8: Acute decompensation"""
        if self.hospitalized:
            return  # Already at hospital

        self.current_day = 8
        self.hospitalized = True
        self.hospitalization_day = 8

        self.patient.add_event(
            day=8,
            event_type="Emergency admission",
            details="Acute decompensated HF, confusion, dyspnea. 24hr delay from fall."
        )

    def calculate_outcome(self):
        """Calculate clinical outcomes"""
        outcomes = {
            'hospitalized': self.hospitalized,
            'hospitalization_day': self.hospitalization_day,
            'los': None,  # length of stay
            'aki': False,
            'readmission_30d': False
        }

        if self.hospitalized:
            # Length of stay depends on timing of intervention
            if self.hospitalization_day == 7:
                # Early intervention: shorter stay
                outcomes['los'] = np.random.choice([2, 3, 4], p=[0.5, 0.3, 0.2])
                outcomes['aki'] = np.random.random() < 0.20  # 20% AKI risk
                outcomes['readmission_30d'] = np.random.random() < 0.15  # 15% readmission
            else:
                # Delayed intervention: longer stay
                outcomes['los'] = np.random.choice([4, 5, 6, 7], p=[0.3, 0.3, 0.3, 0.1])
                outcomes['aki'] = np.random.random() < 0.45  # 45% AKI risk
                outcomes['readmission_30d'] = np.random.random() < 0.30  # 30% readmission

        return outcomes

print("âœ“ Care pathway model defined")
print("=" * 60)

```

```python
# ===========================
# STEP 4: RUN MONTE CARLO SIMULATION
# ===========================

def run_simulation(n_iterations=1000,
                   escalation_protocol=False,
                   weekend_protocol=False,
                   family_education=False):
    """
    Simulate n_iterations of Mr. Hanna's care pathway

    Args:
        n_iterations: Number of simulations to run
        escalation_protocol: Enable risk algorithm on Day 1
        weekend_protocol: Enable standardized fall response
        family_education: Enable family decision support

    Returns:
        DataFrame with outcomes
    """
    results = []

    for i in range(n_iterations):
        # Create fresh instance
        patient = Patient("Hassan Hanna", 78,
                         ['heart_failure', 'diabetes', 'ckd', 'hypertension'])
        pathway = CarePathway(patient)

        # Simulate care journey
        gp_days = pathway.day1_assessment(escalation_protocol)

        if gp_days <= 3:
            # Early GP visit prevents crisis
            outcomes = {
                'scenario': i,
                'hospitalized': False,
                'hospitalization_day': None,
                'los': 0,
                'aki': False,
                'readmission_30d': False
            }
        else:
            # Crisis occurs
            went_to_er = pathway.day7_fall(weekend_protocol, family_education)
            if not went_to_er:
                pathway.day8_crisis()

            outcomes = pathway.calculate_outcome()
            outcomes['scenario'] = i

        results.append(outcomes)

    return pd.DataFrame(results)

# Run baseline simulation (current state)
print("Running BASELINE simulation (1000 iterations)...")
baseline_results = run_simulation(
    n_iterations=1000,
    escalation_protocol=False,
    weekend_protocol=False,
    family_education=False
)

# Run Intervention A simulation
print("Running INTERVENTION A simulation (Risk Algorithm + GP Alerts)...")
intervention_a_results = run_simulation(
    n_iterations=1000,
    escalation_protocol=True,
    weekend_protocol=False,
    family_education=False
)

# Run Intervention B simulation
print("Running INTERVENTION B simulation (Weekend Protocol + Family Ed)...")
intervention_b_results = run_simulation(
    n_iterations=1000,
    escalation_protocol=False,
    weekend_protocol=True,
    family_education=True
)

# Run Combined Interventions
print("Running COMBINED INTERVENTIONS simulation...")
combined_results = run_simulation(
    n_iterations=1000,
    escalation_protocol=True,
    weekend_protocol=True,
    family_education=True
)

print("âœ“ All simulations complete")
print("=" * 60)

```

```python
# ===========================
# STEP 5: ANALYZE RESULTS
# ===========================

def calculate_metrics(df, scenario_name):
    """Calculate key performance metrics"""
    metrics = {
        'Scenario': scenario_name,
        'Hospitalization Rate (%)': (df['hospitalized'].sum() / len(df)) * 100,
        'Average LOS (days)': df[df['hospitalized']]['los'].mean(),
        'AKI Rate (%)': (df['aki'].sum() / len(df)) * 100,
        '30-day Readmission (%)': (df['readmission_30d'].sum() / len(df)) * 100,
    }

    # Calculate early vs late admission
    early_admission = df[df['hospitalization_day'] == 7].shape[0]
    late_admission = df[df['hospitalization_day'] == 8].shape[0]

    if (early_admission + late_admission) > 0:
        metrics['Early Intervention (%)'] = (early_admission / (early_admission + late_admission)) * 100
    else:
        metrics['Early Intervention (%)'] = 0

    return metrics

# Calculate metrics for all scenarios
comparison = pd.DataFrame([
    calculate_metrics(baseline_results, "Baseline (Current State)"),
    calculate_metrics(intervention_a_results, "Intervention A (Risk Algorithm)"),
    calculate_metrics(intervention_b_results, "Intervention B (Weekend Protocol)"),
    calculate_metrics(combined_results, "Combined Interventions")
])

print("\n" + "=" * 60)
print("SIMULATION RESULTS SUMMARY")
print("=" * 60)
print(comparison.to_string(index=False))
print("=" * 60)

```

```python
# ===========================
# STEP 8: SENSITIVITY ANALYSIS
# ===========================

print("\n" + "=" * 60)
print("SENSITIVITY ANALYSIS")
print("=" * 60)
print("Testing robustness of findings across different assumptions...")

# Test different risk thresholds for Intervention A
thresholds = [6, 7, 8, 9, 10]
threshold_results = []

for threshold in thresholds:
    # Modify simulation to test threshold
    temp_results = []
    for i in range(500):
        patient = Patient("Test", 78, ['heart_failure', 'diabetes', 'ckd', 'hypertension'])
        pathway = CarePathway(patient)

        # Modified decision logic
        combined_risk = patient.get_baseline_risk() + np.random.choice([1, 2, 3, 4])
        if combined_risk >= threshold:
            prevented = True
        else:
            prevented = False

        temp_results.append(prevented)

    prevention_rate = (sum(temp_results) / len(temp_results)) * 100
    threshold_results.append({
        'Threshold': threshold,
        'Prevention Rate (%)': prevention_rate
    })

sensitivity_df = pd.DataFrame(threshold_results)
print(sensitivity_df.to_string(index=False))

# Visualization 6: Sensitivity Analysis
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(sensitivity_df['Threshold'], sensitivity_df['Prevention Rate (%)'],
        marker='o', linewidth=2, markersize=10, color='#1f77b4')
ax.axvline(x=8, color='red', linestyle='--', label='Optimal Threshold (8)', linewidth=2)
ax.set_xlabel('Risk Score Threshold', fontsize=12)
ax.set_ylabel('Hospitalization Prevention Rate (%)', fontsize=12)
ax.set_title('Sensitivity Analysis: Risk Stratification Threshold', fontsize=14, fontweight='bold')
ax.grid(alpha=0.3)
ax.legend()
plt.tight_layout()
plt.savefig('/mnt/user-data/outputs/sensitivity_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nâœ“ Visualization 3 saved: sensitivity_analysis.png")



```
